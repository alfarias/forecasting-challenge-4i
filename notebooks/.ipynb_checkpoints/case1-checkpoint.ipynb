```python
import pandas as pd
import numpy as np
import hvplot.pandas
from IPython.display import HTML

data_path = 'D:\\Projects\\forecasting-challenge-4i\\data\\'
```

```python
data = pd.read_csv(data_path+'TFP.csv')
display(data.head(3))
data.info()
```

```python
data.isnull().sum()
```

```python
print(f'Countries: {list(data["isocode"].unique())}')
print(f'Years: {list(data["year"].unique())}')
```

```python
countries_tfp = data.hvplot.line(x='year', y='rtfpna', by='isocode',
                 ylabel='TFP', subplots=True).cols(1)
display(HTML('<span style="font-weight:bold">' + 'TFP from Years 1950-2010 by Countries'\
             + '</span>'), countries_tfp)
```
